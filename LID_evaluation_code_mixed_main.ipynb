```python
for ind in df.index:
     inputText = df['tweet'][ind]
     list = langIdentify(inputText,"classifiers/HiEn.classifier")
     print(list)
     n = len(list[0])
     u = 0
     ci=0
     dict= {'EN': 0, 'HI': 0, 'OTHER': 0} #uni 
     for tuple in list:
         for element in tuple:
             #element_one = element[0]
             element_two = element[-1]
             #print("Element1:",element_one)
             #print("Element2:",element_two)
             dict[element_two]+=1
     all_values = dict.values()
     u = u+dict['OTHER']
     w = max(dict['EN'],dict['HI'])
     print("EN",dict['EN'],"HI",dict['HI'])
     print("w",w,"u",u,"n",n)
     if n>u:
         ci = 100*(1 - w/(n-u))
     if n==u:
         ci=0
     print("CMI Score: ",ci)
```
